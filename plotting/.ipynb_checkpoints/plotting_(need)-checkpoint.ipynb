# 1 A simple plot

## 1.1 Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',              # creating plot and setting parameters of graph, red dashed line can be set as : 'r--'
         color='r', linestyle='dashed')
plt.xlabel('x-values')      # define X axis
plt.ylabel('y-values')      # define y axis
plt.title('X vs Y')         # define title
plt.grid(alpha=.25)         # creates grid in the graph
plt.legend(loc='upper left')  # define location of legend
```

## 1.2 Adding another plot

```python

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]


plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue')  # To plot data of y_2 into the same graph
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 1.3 Yet another plot but with error bars

```python

x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75


plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,         # To add error bar, xerr refers to x error, yerr refers to y error
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75


plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,         # To add error bar, xerr refers to x error, yerr refers to y error
             label='Y3 with errors', color='green',alpha=.5)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# 2 Better with NumPy

```python
# Some data for plotting
import numpy as np
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)         # using numpy arrays to plot graph instead of lists
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

```python
import numpy as np
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)         # using numpy arrays to plot graph instead of lists
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1**2, color='red', linestyle='dashed', label='Y values') # to create a graph containing sqaured values
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # numpy arrays allows us to generate mathematical functions
sin_x = np.sin(x)          

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
```

## We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)


plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # To create new y-axis  for the plots that comes after
                                                   
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

# 3 Saving to disc

```python
import numpy as np
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.jpeg', dpi=150)  #To save a high resolution image of the graph in the folder
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
#filename = 'Complete_TAVG_daily.txt'
#data = np.loadtxt(filename, skiprows=2)
#x=data[:,0]
#y=data[:,1]
#plt.plot(x,y)
```

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)        # to create a numpy array using values from txt file
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-4, 4])
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey') # to create a horizontal line at zero value
```

## 4.2 Styles

```python
plt.style.use('fivethirtyeight') # To use style template fivethirtyeight

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
