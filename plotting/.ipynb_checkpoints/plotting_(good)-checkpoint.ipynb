# 2 Comparing the two ‘dialects’

```python
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

### pyplot Interface

```python
plt.plot(x, cos_x, label='cos x') # Plot cos x and label as 'cosx'
plt.plot(x, sin_x, label='sin x') # Plot sin x and label as 'sinx'
plt.legend()
```

### OO Interface

```python
fig, ax = plt.subplots(nrows=1, ncols=1) # to create figure with one column and one row, can also just use plt.subplots() if just need one column and one row
ax.plot(x, cos_x, label='cos x') # to plot axis for cosx
ax.plot(x, sin_x, label='sin x') # to plot axis for sinx
ax.legend() # create legend
ax.spines[['right','top']].set_visible(False) #To remove border at the top and right
```

# 3 What is OO ax, really?

```python
fig, ax = plt.subplots(nrows=2, ncols=1) # Using plt.subplots() to create a figure with 2 rows and one column

ax[0].plot(x, cos_x, label='cos x', c='r') # To plot cosx in the first axis, set colour of line red
ax[1].plot(x, sin_x, label='sin x') # To plot sinx in the second axis

ax[0].legend() 
ax[1].legend()
```

```python
type(ax) # ax is a numpy array
```

```python
ax.shape # ax currently has 2 columns(2 dimensional array)
```

```python
type(ax[0]) # ax[0] is a Matplotlib axis
```

# 4 A complete OO example

```python
fig, ax = plt.subplots(nrows=2, ncols=1,    # To set 2 axis with 2 rows and 1 column, set the size of the figure(in inches) and share the x axis
                       figsize=(5, 5),
                       sharex=True)
ax[0].plot(x, cos_x, label=r'$\cos(x)$')  # To plot cos x in first axis
ax[0].fill_between(x, 1, cos_x, alpha=.25) # TO fill the outside of the plot
ax[1].plot(x, sin_x, label=r'$\sin(x)$') # To plot sinx in second axis
ax[1].fill_between(x, 0, sin_x, alpha=.25) # To fill the plots

for a in ax: # set a for loop such that it will iterate through all the axis to create a legend for each axis
    a.legend()  
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$') # Label the second plot
ax[1].set_title('$sin(x)$') # set title for the axis

fig.suptitle(r'$\sin(x)$ and $\cos(x)$') # To set title for the figure
fig.tight_layout() # adjusts the layout of the axes within a figure to minimise overlaps of the axes
```

# 5 Multiple rows and columns

```python
fig, ax = plt.subplots(nrows=2, ncols=2,      # Create a figure with 2 rows and 2 columns, subplots in each column should share their x axis and y axis
                       figsize=(5, 5),
                       sharex='col', sharey='col')

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4) # To draw the horizonal line in the top left axis
top_right.hlines(y=2, xmin=0, xmax=5) # To draw the horizonal line in the top right axis
bottom_left.hlines(y=3, xmin=0, xmax=4) # To draw the horizonal line in the bottom left axis
bottom_right.hlines(y=4, xmin=0, xmax=5) # To draw the horizonal line in the bottom right axis

for a in ax.flatten(): # To apply a for loop to set the visibility of grid to 25%
    a.grid(alpha=.25)

plt.tight_layout()
```

sharex = True, sharey=True       # to share all x and y axis

## 5.1 Sharing axes

```python
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5)
)


ax[0,0].hlines(y=1, xmin=0, xmax=5)
ax[0,1].hlines(y=2, xmin=0, xmax=5)
ax[1,0].hlines(y=3, xmin=0, xmax=5)
ax[1,1].hlines(y=4, xmin=0, xmax=5)
```

```python
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5),
    sharex=True, sharey='row'  # make all the axes use the same range for x axis and use the same range for y axis for each row
)

ax[0,0].hlines(y=1, xmin=0, xmax=5)
ax[0,1].hlines(y=2, xmin=0, xmax=5)
ax[1,0].hlines(y=3, xmin=0, xmax=5)
ax[1,1].hlines(y=4, xmin=0, xmax=5)
```

```python
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5),
    sharex='col', sharey='row'     # To share the x axis for the axes in the same column and share the y axis for the axes in the same row
)
ax[0,0].hlines(y=1, xmin=0, xmax=5)
ax[0,1].hlines(y=2, xmin=0, xmax=5)
ax[1,0].hlines(y=3, xmin=0, xmax=5)
ax[1,1].hlines(y=4, xmin=0, xmax=5)
```

# 6 Other useful plots

### Data

```python
def get_plot_data():
    data = {}
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    data['All'] = np.loadtxt(filename, skiprows=1)

    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    data['NUS'] = np.loadtxt(filename, skiprows=1)

    return data
```

## 6.1 Histograms

```python
plt.style.use('bmh') # To set specific style
data = get_plot_data()

# bins specifies how many bins to split the data
plt.hist([data['All'], data['NUS']], bins=50, label=['All', 'NUS']) #bins = : to set the number of bins
plt.xlabel('Mean of Basic Montly Earning (S$)')
plt.ylabel('Number of Students')
plt.legend()
```

if bins is not specified, matpltlib will set a default number of bins


## 6.2 Scatter plots

```python
plt.style.use("seaborn-v0_8-darkgrid") # To set specific style

data = get_plot_data()

for plot_label, numbers in data.items():
    x = range(len(numbers))   
    y = numbers
    plt.scatter(x, y, label=plot_label, alpha=.5) 

plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Montly Eraning (S$)')
plt.legend()
```

## 6.3 Bar charts

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys()) # convert the key in student_numbers dictionary into a list
numbers = list(student_numbers.values()) # convert the values in student_numbers dictionary into a list

plt.style.use('ggplot') # To set specific style
plt.bar(majors, numbers, color =['r','g','y','b']) # to create a bar chart with x axis as major and y axis as number of students and setting specific colour for each bar
plt.xlabel('Majors')
plt.ylabel('Number of Students')
```

## 6.4 Pie charts

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('fivethirtyeight')
plt.pie(numbers, 
        labels=majors,
        autopct='%1.1f%%',   # To format the percentages
        startangle=-90                
        )
plt.title('Percentage of each major')
```
