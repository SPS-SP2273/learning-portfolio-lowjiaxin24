```python
import os
import shutil
import glob
```

**Task 1**

```python
legend = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

path = os.path.join('os-collaboration-exercise-data', 'oct-08_singapore.txt')
filename = path.split(os.path.sep)[-1]    # Take filename
month_only = filename.split('-')[0]       # Split the filename and take the month only
daywcountry_only = filename.split('-')[1]  #Split the filename and take the day_country.txt
day_only = daywcountry_only.split('_')[0]  #Split day_country.txt to get day
countrywfiletype_only = filename.split('_')[-1]  # Split filename to get country.txt

for key, value in legend.items(): 
    if month_only == key:     
        month_only = value    # if month_only == key, replace month_only with value

new_file_name = f'{month_only}-{day_only}-{countrywfiletype_only}'   # new file name as month_only-day_only-countrywfile_only
print(f'{new_file_name}')

    
    

 

```

**Task 2**

```python
def rename_my_file(old_file_name): #define rename_my_file function using old_file_name as the input
    old_file_name = input('old filename:')  # Obtain for input of old file name
    path = os.path.join('os-collaboration-exercise-data(1)', 'os-collaboration-exercise-data', old_file_name) 
    filename = path.split(os.path.sep)[-1]   
    month_only = filename.split('-')[0]       
    daywcountry_only = filename.split('-')[1]  
    day_only = daywcountry_only.split('_')[0]
    countrywfiletype_only = filename.split('_')[-1]

    for key, value in legend.items():
        if month_only == key:
            month_only = value
    
    new_file_name = f'{month_only}-{day_only}-{countrywfiletype_only}'
    return new_file_name          # return new_file_name
```

**Task 3**

```python
file_to_replace = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']

def rename_my_file(file):                                             # define rename_my_file function using file as the input
    path = os.path.join('os-collaboration-exercise-data', file)
    filename = path.split(os.path.sep)[-1]    
    month_only = filename.split('-')[0]       
    daywcountry_only = filename.split('-')[1]  
    day_only = daywcountry_only.split('_')[0]
    countrywfiletype_only = filename.split('_')[-1]

    for key, value in legend.items():
        if month_only == key:
            month_only = value
    
    new_file_name = f'{month_only}-{day_only}-{countrywfiletype_only}'
    return new_file_name

for file in file_to_replace:            # looping through each variable in the list 'file_to_replace'
    new = rename_my_file(file)          # obtain new filename using rename_my_file function and variables from file_to_replace list 
    print(f'{file} ----> {new}')        
```

**Task 4**

```python
search_txt = glob.glob('os-collaboration-exercise-data/*.txt') # creating a path that search for all the .txt files in os-collaboration-exercise data folder

for file in search_txt:                       # looping through all the paths 
    filename = os.path.split(file)[1]         # Obtain the filename from all the paths
    new_Filename = rename_my_file(filename)   # Obtain new filename using rename_my_file function and variables from filename from all the paths
    os.rename(file, f'os-collaboration-exercise-data\\{new_Filename}') # rename all the .txt files in the os-collaboration-exercise data folder
```

**Task 5**

```python
list_of_cities=['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in list_of_cities:
    path = os.path.join('os-collaboration-exercise-data', city) # create a path 'os-collaboration-exercise-data\\city'
    print(f'Creating {path}')
    os.mkdir(path)          # create a folder for each city using each path
```

**Task 6**

```python
for cairo_file in glob.glob('os-collaboration-exercise-data\\*cairo*'):  # looping through all the files with 'cairo' in the filename
    filename = os.path.split(cairo_file)[1]                              # obtain the filename from the path
    path_to_destination = os.path.join('os-collaboration-exercise-data', 'Cairo')  # create path to folder to be copied to
    source_path = os.path.join('os-collaboration-exercise-data', filename)   # create path of file to be copied
    shutil.copy(source_path, path_to_destination)   # copy files to the folder
    print(f'Copied {filename} to {path_to_destination}') 
    
```

**Task 7**

```python
list_of_cities=['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in list_of_cities:           # looping through all the variables in the list_of_cities list
    for city_file in glob.glob(f'os-collaboration-exercise-data\\*{city}*'):     # looping through all the files with the city name in the filename
        filename = os.path.split(city_file)[1]  # obtain the filename from the path
        path_to_destination = os.path.join('os-collaboration-exercise-data', city) # create path to folder to be copied to
        source_path = os.path.join('os-collaboration-exercise-data', filename) # create path of file to be copied
        shutil.move(source_path, path_to_destination)  # move file to the folder
        print(f'Moved {filename} to {path_to_destination}')
```
