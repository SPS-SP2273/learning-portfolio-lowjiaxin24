### Writing to a file in one go

```python
with open('my-text-once.txt', 'w') as file:    # writes the text as one file 'my-text-once.txt' that is saved in the desktop
    file.write(text)
```

If this code runs again, the existing file will be overwritten

### Writing to a file, line by line

```python
with open('my-text-lines.txt', 'w') as file:  # writes the file line by line
    for line in text.splitlines():
        file.writelines(line)
```

# 5 Folders

## 5.1 Creating folders

```python
os.mkdir('people') # create a folder 'people'

for person in ['John', 'Paul', 'Ringo']: # create files 'John','Paul','Ringo' in 'person' folder using a loop
    path = os.path.join('people', person) 
    print(f'Creating {path}')    # Dont need to include print() -  just to show what is happening
    os.mkdir(path) # To create files 'John','Paul','Ringo' into 'people' folder which is in person folder
```

## 5.2 Checking for existence

### Using try-except

```python
for person in ['John', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    try:                        # To create files 'John','Paul','Ringo' into 'people' folder which is in person folder if the files are not created
        print(f'Creating {path}')
    except FileExistsError:    # If files John','Paul','Ringo' are present, skip creating the files and print message '{path} already exists;skipping creation' to notify user that files are already created
        print(f'{path} already exists; skipping creation.')
```

### Using os.path.exists()

```python
for person in ['John', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    if os.path.exists(path):            # If files John','Paul','Ringo' are already present,  print message '{path} already exists;skipping creation' to notify user that files are already created
        print(f'{path} already exists; skipping creation.')
    else:                              # Else, create files 'John','Paul','Ringo' into 'people' folder
        os.mkdir(path)
        print(f'Creating {path}')
```

## 5.3 Copying files

```python
file_to_copy = 'sp2273_logo.png'
for person in ['John', 'Paul', 'Ringo']:
    path_to_destination = os.path.join('people', person) # To create path of each person file in the 'people' directory
    file_path = os.path.join(path_to_destination, file_to_copy) # To create path by joining the destination directory to 'sp2273_logo.png' file
    if os.path.exists(file_path):
        print(f'{file_path} already exists!')  # If file_path is present, print message 'file_path already exists'
    else:        
        shutil.copy('sp2273_logo.png', path_to_destination)  # Else, copy 'sp2273_logo.png' file to each person file in the 'people' directory
        print(f'Copied {file_to_copy} to {path_to_destination}')
```

```python
file_to_copy='sp2273_logo.png'
for person in ['John', 'Paul', 'Ringo']:
    # Create folder 'imgs'
    path_to_imgs = os.path.join('people', person, 'imgs') 
    if not os.path.exists(path_to_imgs):
        os.mkdir(path_to_imgs)

    # Move logo file
    current_path_of_logo = os.path.join('people', person, file_to_copy)
    new_path_of_logo = os.path.join('people', person, 'imgs', file_to_copy)

    shutil.move(current_path_of_logo, new_path_of_logo) # move file from current_path_of_logo to new_path_of logo
    print(f'Moved logo from {current_path_of_logo}to {new_path_of_logo}')
```

# 6 Listing and looking for files

```python
glob.glob('*') # To read and tell you all the files that are present in the current directory
```

```python
glob.glob('peo*') # To find and tell you the files in which their name contain 'peo' at the beginning 
```

```python
glob.glob('peo*/*') # To tell you what is present in the folders that start with 'peo'
```

```python
glob.glob('people/**', recursive=True) # To tell you the sub-directories that are present in folders that start with 'peo'
```

```python
glob.glob('people/**/*.png', recursive=True) # To tell you the .png files that are in the people folder
```
