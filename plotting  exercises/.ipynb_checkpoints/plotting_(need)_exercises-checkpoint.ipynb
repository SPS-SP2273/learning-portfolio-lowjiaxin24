```python
from matplotlib import pyplot as plt
```

**Exercise 1**

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr= x_err, yerr= y_err,
             color= 'brown', linestyle= 'dashed',
             marker= 'o', markerfacecolor='yellow',
             ecolor='black',capsize=5,
             label= 'Y values with errors')   # To create brown dashed error line with yellow circle markers and black error bar caps, labelled as 'Y values with errors'
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc= 'lower right') # To place legend at lower right
plt.show()
```

**Exercise 2**

```python
import numpy as np
```

```python
np_x_values = np.linspace(0,50,200)   # to generate 200 x values from 0 to 50 that are equally spaced
np_y_values = np.exp(-np_x_values / 10) * np.sin(np_x_values) # to calculate corresponding y values using the 200 x values
```

```python
plt.plot(np_x_values, np_y_values, 
             color= 'steelblue')   # To plot steelblue line with np_x_values as the x values and np_y_values as the y values
plt.xlabel('x', color= 'darkorange', fontsize = 15) # to create dark orange x axis label 
plt.ylabel('y', color= 'darkorange', fontsize = 15, rotation=0) # to create dark orange y axis label
plt.title('y = $e^{-x/10}sin(x)$', fontsize = 20) # to create title with font size of 20 
plt.grid(alpha= 0.25) # to set opacity of the grid to 25%
plt.ylim(-1,1) # to set the limits of the y axis from -1 to 1
plt.show()
```
